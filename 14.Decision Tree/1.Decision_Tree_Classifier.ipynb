```python
import pandas as pd
import numpy as np
import seaborn as  sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
#independent features
X=pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width','petal length','petal width'])
```

```python
#dependent feature
y=iris['target']
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
#Decision does not require any standardization or any kind of normalization

#apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier
tree_classifier = DecisionTreeClassifier()
```

```python
tree_classifier.fit(X_train,y_train)
```

```python
#Visualize the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_classifier,filled=True)
```

```python
#Prunning (Max_depth)
#apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier
tree_classifier1 = DecisionTreeClassifier(max_depth=2)
```

```python
tree_classifier1.fit(X_train,y_train)
```

```python
#Visualize the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_classifier1,filled=True)
```

```python
y_pred=tree_classifier.predict(X_test)
y_pred1 = tree_classifier1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
cm = confusion_matrix(y_test,y_pred)
cm1 = confusion_matrix(y_test,y_pred1)

print(cm)
print(cm1)

print(classification_report(y_test,y_pred))
print(classification_report(y_test,y_pred1))

```

```python
#Pre Prunning

params = {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
params
```

```python
#Hyperparameter tunning
from sklearn.model_selection import GridSearchCV
tree_model = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(tree_model,param_grid=params,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred2 = grid.predict(X_test)
```

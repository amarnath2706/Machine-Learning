```python
import pandas as pd
import numpy as np
import seaborn as  sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
diab = load_diabetes()
```

```python
print(diab['DESCR'])
```

```python
import pandas as pd
df_diab= pd.DataFrame(diab.data,columns=['age',
                                         'sex',
                                         'bmi',
                                         'bp',
                                         's1',
                                         's2',
                                         's3',
                                         's4',
                                         's5',
                                         's6'])
df_diab.head()
```

```python
#independent and dependent feature
X=df_diab
y=diab['target']
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
 #Check correlation
X_train.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
#Decisontree regressor
from sklearn.tree import DecisionTreeRegressor
tree_regressor = DecisionTreeRegressor()
tree_regressor.fit(X_train,y_train)
```

```python
#Visualize the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_regressor,filled=True)
```

```python
y_pred=tree_regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
#Hyperparameter Tunning
params = {
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
params
```

```python
#Hyperparameter tunning
from sklearn.model_selection import GridSearchCV
tree_model_regressor = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(tree_model_regressor,
                    param_grid=params,
                    cv=5,
                    scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
#prediction
y_pred=tree_regressor.predict(X_test)
y_pred1=grid.predict(X_test)
```

```python
print(r2_score(y_test,y_pred1))
print(mean_absolute_error(y_test,y_pred1))
print(mean_squared_error(y_test,y_pred1))
```

```python
selected_params_model = DecisionTreeRegressor(criterion='friedman_mse',
                                              max_depth=3,
                                              max_features='sqrt',
                                              splitter='best'
                                              )
```

```python
selected_params_model.fit(X_train,y_train)
```

```python
#Visualize the decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selected_params_model,filled=True)
```

```python
y_pred_selected_model=selected_params_model.predict(X_test)
```

```python
print(r2_score(y_test,y_pred_selected_model))
print(mean_absolute_error(y_test,y_pred_selected_model))
print(mean_squared_error(y_test,y_pred_selected_model))
```

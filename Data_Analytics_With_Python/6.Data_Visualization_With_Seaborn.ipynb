```python
import seaborn as sns
```

```python
#Basic plots
tips=sns.load_dataset('tips')
tips
```

```python
#Scatterplot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Tips based on total bill")
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,color='teal')
plt.title("Tips based on total bill",color='teal')
plt.show()
```

```python
#Line plot
sns.lineplot(x='size',y='total_bill',data=tips,color='teal')
plt.title("Tips based on total bill",color='teal')
plt.show()
```

```python
#Categorical plot
#Bar plot
sns.barplot(x='day',y='total_bill',data=tips,color='teal')
plt.title("Tips based on total bill",color='teal')
plt.show()
```

```python
#Box plot
sns.boxplot(x='day',y='total_bill',data=tips,color='teal')
plt.title("Tips based on total bill",color='teal')
plt.show()
```

```python
#Violin plot - helpful to see the distribution of the data
sns.violinplot(x='day',y='total_bill',data=tips,color='teal')
plt.title("Tips based on total bill",color='teal')
plt.show()
```

```python
#Histogram
sns.histplot(x='day',y='total_bill',data=tips,color='teal')
plt.title("Tips based on total bill",color='teal')
plt.show()
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True,color='Teal')
plt.title("total bill",color='teal')
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=False,color='Teal')
plt.title("total bill",color='teal')
```

```python
#kde plot
sns.kdeplot(tips['total_bill'],shade=True,color='teal')
```

```python
sns.kdeplot(tips['total_bill'],fill=True,color='teal')
```

```python
#Pairplot
sns.pairplot(tips)
```

```python
#Heat map
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='mako')
```

```python
import pandas as pd
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
#total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category', y='Total Revenue',data=sales_df,estimator=sum,color='teal')
plt.title('Total sales by product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Total Revenue',data=sales_df,estimator=sum,color='teal')
plt.title('Total sales by product')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

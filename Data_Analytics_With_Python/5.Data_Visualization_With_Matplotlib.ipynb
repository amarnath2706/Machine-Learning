```python
#Create multiple plots using subplots
x1=[1,2,3,4,5]
y1=[1,4,9,16,25]
z1=[1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1) #here iam going to plot 1 row 2 columns
plt.plot(x1,y1, color='darkviolet')
plt.title('Multiplots_Plot1')

plt.subplot(1,2,2) #here iam going to plot 1 row 2 columns
plt.plot(y1,z1, color='brown')
plt.title('Multiplots_Plot2')


```

```python
#Create multiple plots using subplots
x1=[1,2,3,4,5]
y1=[1,4,9,16,25]
z1=[1,2,3,4,5]

plt.figure(figsize=(10,8))
plt.subplot(2,2,1) #here iam going to plot 1 row 2 columns
plt.plot(x1,y1, color='darkviolet')
plt.title('Multiplots_Plot1')

plt.subplot(2,2,2) #here iam going to plot 1 row 2 columns
plt.plot(y1,z1, color='brown')
plt.title('Multiplots_Plot2')

plt.subplot(2,2,3) #here iam going to plot 1 row 2 columns
plt.plot(y1,z1, color='green')
plt.title('Multiplots_Plot3')

plt.subplot(2,2,4) #here iam going to plot 1 row 2 columns
plt.plot(x1,z1, color='blue')
plt.title('Multiplots_Plot4')
```

```python
# Piechart
labels=['A','B','C','D','E','F']
sizes = [10,30,40,20,25,15]
colors = ['red','yellowgreen','lightskyblue','lightcoral','pink','gold']
explode = (0.2,0,0,0,0,0)

plt.pie(sizes,labels=labels)
```

```python
plt.pie(sizes,labels=labels,colors=colors)
```

```python
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%")
```

```python
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
#Sales Data Visualization
import pandas as pd 
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
sales_df.info()
```

```python
#based on product category we can find the insights
#plot total sales by products

total_sales_by_products = sales_df.groupby('Product Category')['Total Revenue'].sum()
print(total_sales_by_products)
```

```python
#Bar plot
total_sales_by_products.plot(kind='bar',color='teal')
```

```python
sales_trend = sales_df.groupby('Date')['Total Revenue'].sum()
sales_trend
```

```python
sales_trend = sales_df.groupby('Date')['Total Revenue'].sum().reset_index()
sales_trend
```

```python
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```

```python
total_sales_by_products.plot(kind='pie')
```

```python
colors1 = ['red','lightcoral','blue','gold','teal']
#explode = (0,0,0.2,0,0)
total_sales_by_products.plot(kind='pie',colors=colors1,autopct="%1.1f%%",shadow=True)
```
